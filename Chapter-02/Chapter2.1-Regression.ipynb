**Supervised Learning - Linear Regression**

A set of grades and hours of study is provided and we are estimate the grades given a students effort in terms of hours of study.

This is a Regression problem and we will use a Linear Regression model to predict.

**Import all the required libraries**

```python
# Importing Standard Libraries
import numpy as np
import matplotlib.pyplot as plt

# Importing sklearn Libraries
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.linear_model import LinearRegression
#from sklearn import linear_model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

**In this first example, we are setting the values for Hours of Study & the Grades achieved.**

X is the input variable or the features <br>
y is the dependent variable or the label


```python
x = np.array([8,10,6,7,4,8,7]).reshape(-1,1)
y = [37,49,35,38,30,45,37]
```

**In the next cell, we are viewing a plot of grades(y) to hours of study(X) using matplotlib library. This is our training dataset**

```python
plt.title('Scatter plot of Grades vs Hours of Study per week')
plt.xlabel('Hours of Study per week')
plt.ylabel('Grades')
plt.scatter(x, y, c='b');
```

**We will use the LinearRegression class from sklearn, instantiating the LinearRegression class**

```python
# Instantiate LinearRegression Model
lr = LinearRegression()
```

**The model is trained in the next cell, we use the fit function from the LinearRegression class. The model is fitted on the features (X) and labels(y) and finds the linear model which closes aligns to our provided training dataset.**

```python
# Training/Fitting the Model
model = lr.fit(x, y)
```

**We then use the trained model to run predictions on unseen data.**

```python
xfit = np.linspace(0, 10, 10)                                # create 10 points between 0 and 10
yfit = model.predict(xfit[:, np.newaxis])                      # predict the values for dependent variable 
```

**Next we plot the regression decision line as defined by the trained model. The blue dots are the predicted points and the visual provides an indication of the alignment of the predicted values to the decision line.**

```python
plt.title('Linear Regression decision line for predicting grades')
plt.xlabel('Hours of Study per week')
plt.ylabel('Grades')
plt.scatter(x, y, c='b')
plt.plot(xfit, yfit, 'k');
```

**Next we print the intercept/bias value and the coefficients from the linear regression expression (y=mx+c)**

The values below are indicative and slight changes in the values can be expected depending on the training runs.

```python
print("c=",model.intercept_)
```

```python
print("m=",model.coef_)
```

**This the trained Linear Regression model where the bias and coeeficient values were determined as a result of the training. <br>
y= mx + c**

```python
print("y=",np.round(model.coef_[0],3),"*x","+",np.round(model.intercept_,3))
```

The coefficients 3.13 gives us the magnitude of impact it has on the feature providing us with the ability to interpret the feature's impact on the prediction. <br>
This example has one feature but with more features, the coefficients are mode informative in understanding the models.
